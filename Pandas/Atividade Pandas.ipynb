* PROBLEMA: Encontrar a quantidade minima e horarios que todos conseguem se encontrar

```python
import pandas as pd
from typing import Dict, List

if __name__ == "__main__":
    
    df = pd.read_csv("doodle.csv")
    
    alunos_por_dia: Dict = {}
    #chave: horario; valor: quantidade de alunos que podem participar
    for coluna in df.columns:
        alunos_por_dia[coluna] = df[coluna].count()
    
    alunos_por_dia_decresc: list = sorted(alunos_por_dia, key = alunos_por_dia.get, reverse=True)[1:]
    #ordenado do horario com maior quantidade de alunos para o com menor quantidade
    #exclue o primeiro elemento (str: "Nome do aluno")
    
    alunos_1: list = []
    #alunos que já participaram de pelo menos uma aula
    
    alunos_2: Dict = {}
    #chave: indice do aluno; valor: nome do aluno
    for indice, fileira in df.iterrows():
        alunos_2[indice] = fileira['Nome do aluno']

    horarios: list = []
    #horarios em que pelo menos um aluno que ainda não participou de uma aula pode participar
    
    aluno_adc: bool = False
    #verifica se na iteração houve um aluno que ainda não participou de uma aula
    
    for dia in alunos_por_dia_decresc:
        for indice in range(len(df[dia])):
            if not (alunos_2[indice] in alunos_1):
                if not (df[dia].isnull()[indice]):
                    alunos_1.append(alunos_2[indice])
                    aluno_adc = True
        if aluno_adc:
            horarios.append(dia)
            aluno_adc = False
        if len(alunos_1) >= len(alunos_2):
            break
        
    print("Numero de Encontros (minimo): " + str(len(horarios)))
    print("Horarios_")
    for indice, horario in enumerate(horarios):
        print(f"{indice+1} : {horario}")
```
