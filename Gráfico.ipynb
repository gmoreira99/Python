```python
import matplotlib.pyplot as plt
import numpy as np
from typing import List

def plotGraph(f, x_dados:np.array,limit_inf:float=-10, limit_sup:float=10):
    #f:function
    
    #create 100 equally spaced points between limit_inf and limit_sup
    x = np.linspace(limit_inf, limit_sup, 100)
    #calculate the y value for each element of the x vector
    y:List = []
    for xi in x:
        y.append(f(xi))
        
    plt.plot(x, y)
    
    plt.grid(b=True)
    for xi in x_dados:
        plt.scatter(xi, f(xi), marker="o", color = 'blue')
```

```python
f = lambda x : 0.132+1.56*10**-4*x+2.64*10**-7*x**2
x = np.array([-100, 200], np.int64)
```

```python
plotGraph(f, x, limit_inf=-500, limit_sup=205)
```
