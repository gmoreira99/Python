```python
import numpy as np
from scipy.interpolate import lagrange

X:np.array = np.array([0, 5, 10, 20, 30, 40], np.int32)
Y:np.array = np.array([1.787, 1.519, 1.307, 1.002, 0.7975, 0.6529], np.float)
x_aprox:float = 7.5
y_aprox:float = 0.0

poly:np.poly1d = lagrange(X, Y)

print("Polinomio: ")
print(f"{pol[0]}", end=" ")
if(pol[1] > 0):
        print(f"+ {pol[1]}*X", end=" ")
else:
        print(f"{pol[1]}*X", end=" ")
        
for i in range(2, len(pol)+1):
    if(pol[i] > 0):
        print(f"+ {pol[i]}*X^{i}", end=" ")
    else:
        print(f"{pol[i]}*X^{i}", end=" ")
        
for i in range(len(pol)+1):
    y_aprox += pol[i]*(x_aprox**i)
print(f"\n\nResultado: {y_aprox}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import List

def plotGraph(f, pontos:List, limit_inf:float, limit_sup:float):
    #f:function
    
    #create 100 equally spaced points between limit_inf and limit_sup
    x:np.array = np.linspace(limit_inf, limit_sup, 100)
    #calculate the y value for each element of the x vector
    y:List = []
    for xi in x:
        y.append(f(xi))
        
    plt.plot(x, y)
    
    plt.grid(b=True)
    for xi in pontos:
        plt.scatter(xi, f(xi), marker="o")
    
dados:List = list(X)
dados.append(x_aprox)
f = lambda x : 1.787 - 0.0601865476190476*x + 0.0014248988095237952*(x**2) - 2.2433333333334588e-05*(x**3) + 1.8529761904765798e-07*(x**4) - 4.4047619047657587e-10*(x**5)
plotGraph(f, dados, -10, 50)
```
